## Lending Club Loan Data Processing and Analysis in Python


## 1. Objectives
This module aims at addressing one of the most critical challenges faced by credit risk analysts and various bankers day-to-day i.e. reducing default/bad loans. Bad loans are a huge overhead to all the financial institutions. According to a survey in 2015-2016, more than 16% of all the loans in United States go default, which not only reduces the profit margin for financial institution but sometimes causes huge losses for them. For this purpose, we will be analyzing the data shared by one such Financial Institution, Lending Club, developing insights based on analysis and providing recommendations to reduce default/bad loans.

## 2. Data Description
Lending club, known as the worldâ€™s largest marketplace, provides a platform to connect borrowers and investors. From the statistics section of their website(https://lendingclub.com), we can download publicly available csv files and load them into the python program for analysis and visualization. These files contain complete loan data for all loans issued through the 2007-2015, including the current loan status (Current, Late, Fully Paid, etc.) and latest payment information. The file containing loan data through the "present" contains complete loan data for all loans issued through the previous completed calendar quarter. Additional features include credit scores, number of finance inquiries, address including zip codes, and state, and collections among others.

## 3. Direction of Analysis
For our last deliverable our direction of analysis was as follows:

1. **Data Cleaning and Processing**: Dealing with missing values, data transformation, dimensionality reduction, data visualization, deriving metrics, univariate analysis, bivariate analysis

    - ___Univariate Analysis___: For continuous variables, visualize and measure the central tendency and spread of the variable using various statistical metrics visualization methods. For categorical variables, measure distribution of each category based on frequency. 
    - ___Bivariate/Multivariate Analysis___: Depicting and measuring relationship between two or more variables using visualization and other statistical measurement techniques.

3. **Predictive Analysis**:  Implementing Random Forest and Logistic Regression Models to predict the probability of customers that will default a loan or not.

```python
import warnings
warnings.filterwarnings('ignore') # if there are any warning due to version mismatch, it will be ignored
import os
import pandas as pd
import numpy as np
from sklearn import preprocessing,metrics 
from IPython.core.display import HTML
pd.set_option("display.max_columns",75)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import linear_model,svm
from sklearn.metrics import average_precision_score
from sklearn.metrics import precision_recall_curve


# Loading the data and checking
df_loan = pd.read_csv("LoanStats_2016Q1.csv",low_memory = False)

print(df_loan.shape)
df_loan.isnull().sum()

# Displaying bar chart of columns and count of NA where NA values are greater than 25%
Col_NA = df_loan.isnull().sum()

Col_NA = Col_NA[Col_NA.values >(0.25*len(df_loan))] 
plt.figure(figsize = (20,4))
Col_NA.plot(kind = 'bar')
plt.title('List of Columns & NA counts where NA values are more than 25%')
plt.show()

# Deleting the columns with more than 25% of values as NA
df_loan_cleaned = df_loan
for index,val in Col_NA.iteritems():
  del df_loan_cleaned[index]

# Removing columns where the count of unique value is one
unique = df_loan_cleaned.nunique()
unique = unique[unique.values == 1]
drop_unique = df_loan_cleaned.drop(labels = list(unique.index), axis = 1, inplace = True)
print("Count of columns removed whose count of unique value is one :", drop_unique)

# For emp_length : Replace n/a by 'Self-Employed' and nan by 0
# There are some values in emp_term which are 'n/a', 
# We assume that are 'self-employed'. Because for 'self-employed' applicants, emp-lenth is 'Not Applicable'
print(df_loan_cleaned.emp_length.unique())
df_loan_cleaned.emp_length.fillna('0',inplace = True)
df_loan_cleaned.emp_length.replace(['n/a'],'Self-Employed',inplace = True)
print(df_loan_cleaned.emp_length.unique())

# Remove irrelevant columns
del df_loan_cleaned["zip_code"]


# Removing % sign from int_rate and replacing NaN values in the required continuos with 0
df_loan_cleaned["int_rate"] = df_loan_cleaned['int_rate'].str.replace('%', '')
df_loan_cleaned["int_rate"].fillna(0, inplace = True)
df_loan_cleaned["loan_amnt"].fillna(0, inplace = True)
df_loan_cleaned["funded_amnt"].fillna(0, inplace = True)
df_loan_cleaned["funded_amnt_inv"].fillna(0, inplace = True)
df_loan_cleaned["installment"].fillna(0, inplace = True)
df_loan_cleaned["annual_inc"].fillna(0, inplace = True)
df_loan_cleaned["dti"].fillna(0, inplace = True)

#Cast all continuos variables to numeric
numeric_columns = ['loan_amnt','funded_amnt','funded_amnt_inv','installment','int_rate','annual_inc','dti']
df_loan_cleaned[numeric_columns] = df_loan_cleaned[numeric_columns].apply(pd.to_numeric)
df_loan_cleaned.tail(3)
```

```python
# Loan amount to Annual Income ratio
df_loan_cleaned.drop(df_loan_cleaned.tail(4).index,inplace = True)
ratio1 = df_loan_cleaned['loan_income_ratio'] = df_loan_cleaned['loan_amnt']/df_loan_cleaned['annual_inc']
print('Loan amount to Annual Income ratio\n', ratio1)
```

```python
# Extracting Year & Month from Issue date
df_loan_cleaned['issue_month'], df_loan_cleaned['issue_year'] = df_loan_cleaned['issue_d'].str.split('-', 1).str
print(df_loan_cleaned[['issue_d','issue_month','issue_year']].head())

# Changing order of months from Jan to Dec, currently it's in alphabetical order(A-Z)
months_order = ["Jan", "Feb", "Mar", "Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
df_loan_cleaned['issue_month'] = pd.Categorical(df_loan_cleaned['issue_month'], categories = months_order, ordered=True)

# Creating Bins for range of Loan Amount
bins = [0, 5000, 10000, 15000, 20000, 25000,40000]
slot = ['0-5000', '5000-10000', '10000-15000', '15000-20000', '20000-25000','25000 and above']
df_loan_cleaned['loan_amnt_range'] = pd.cut(df_loan_cleaned['loan_amnt'], bins, labels = slot)

# Creating Bins for range of Annual Income
bins = [0, 25000, 50000, 75000, 100000,1000000]
slot = ['0-25000', '25000-50000', '50000-75000', '75000-100000', '100000 and above']
df_loan_cleaned['annual_inc_range'] = pd.cut(df_loan_cleaned['annual_inc'], bins, labels = slot)

# Creating Bins for range of Interest rates
bins = [0, 7.5, 10, 12.5, 15,20]
slot = ['0-7.5', '7.5-10', '10-12.5', '12.5-15', '15 and above']
df_loan_cleaned['int_rate_range'] = pd.cut(df_loan_cleaned['int_rate'], bins, labels = slot)
```

```python
def univariate(df,col,vartype,hue =None):
    
    '''
    Univariate function will plot the graphs based on the parameters.
    df      : dataframe name
    col     : Column name
    vartype : variable type : continuos or categorical
                Continuos(0)   : Distribution, Violin & Boxplot will be plotted.
                Categorical(1) : Countplot will be plotted.
    hue     : It's only applicable for categorical analysis.
    
    '''
    sns.set(style = "darkgrid")
    
    if vartype == 0:
        fig, ax = plt.subplots(nrows = 1,ncols = 3,figsize = (20,8))
        ax[0].set_title("Distribution Plot")
        sns.distplot(df[col],ax = ax[0])
        ax[1].set_title("Violin Plot")
        sns.violinplot(data = df, x = col,ax = ax[1], inner = "quartile")
        ax[2].set_title("Box Plot")
        sns.boxplot(data = df, x = col,ax = ax[2],orient = 'v')
    
    if vartype == 1:
        temp = pd.Series(data = hue)
        fig, ax = plt.subplots()
        width = len(df[col].unique()) + 6 + 4*len(temp.unique())
        fig.set_size_inches(width , 7)
        ax = sns.countplot(data = df, x = col, order = df[col].value_counts().index,hue = hue) 
        if len(temp.unique()) > 0:
            for p in ax.patches:
                ax.annotate('{:1.1f}%'.format((p.get_height()*100)/float(len(df_loan))), (p.get_x()+0.05, p.get_height()+20))  
        else:
            for p in ax.patches:
                ax.annotate(p.get_height(), (p.get_x()+0.32, p.get_height()+20)) 
        del temp
    else:
        exit
        
    plt.show()

# Analysis of Loan Amount
univariate(df = df_loan_cleaned,col = 'loan_amnt',vartype=0)

# Analysis of Interest Rate
univariate(df = df_loan_cleaned,col = 'int_rate',vartype=0)

# Annual Income
df_loan_cleaned["annual_inc"].describe()

# Max value is 9500000 which is approx 150 times more than mean value, so we will remove the outliers from Annual Income.
# Remove Outliers (values from 99 to 100%)
q = df_loan_cleaned["annual_inc"].quantile(0.995)
df_loan_cleaned = df_loan_cleaned[df_loan["annual_inc"] < q]
df_loan_cleaned["annual_inc"].describe()
univariate(df = df_loan_cleaned,col = 'annual_inc',vartype = 0)

print("Analysis of Loan Status")
univariate(df = df_loan_cleaned,col = 'loan_status',vartype = 1)

print("Analysis of Home Ownership wise loan")
df_loan_cleaned.home_ownership.unique()

# Remove rows where home_ownership'=='OTHER', 'NONE', 'ANY', nan
rem = ['OTHER', 'NONE', 'ANY']
df_loan_cleaned.drop(df_loan_cleaned[df_loan['home_ownership'].isin(rem)].index,inplace = True)
df_loan_cleaned.home_ownership.unique()
univariate(df = df_loan_cleaned,col = 'home_ownership',vartype = 1,hue = 'loan_status')

print("Analysis of Loan Term")
univariate(df = df_loan_cleaned,col = 'term',vartype = 1,hue = 'loan_status')
```

## 3.2.e Why or when does charge off occur? Which variables are strongly/weakly correlated with probability of charge off?

```python
# UDF: crosstab
# 'crosstab' function will be used to get the summarized table for each column(passed as a variable) 
# along with probability of charge off w.r.t that column.
def crosstab(df,col):
    '''
    df : Dataframe
    col: Column Name
    '''
    crosstab = pd.crosstab(df[col], df['loan_status'],margins=True)
    crosstab['Probability_Charged Off'] = round((crosstab['Charged Off']/crosstab['All']),3)
    crosstab = crosstab[0:-1]
    return crosstab

# UDF: bivariate_prob
# bivariate_prob' function will be used to plot count of values for each column(passed as a variable) 
# stacked across 'loan_status' along with probability of charged off as a line chart.
# Probability of charge off
def bivariate_prob(df,col,stacked= True):
    '''
    df      : Dataframe
    col     : Column Name
    stacked : True(default) for Stacked Bar
    '''
    # get dataframe from crosstab function
    plotCrosstab = crosstab(df,col)
    
    linePlot = plotCrosstab[['Probability_Charged Off']]      
    barPlot =  plotCrosstab.iloc[:,0:2]
    ax = linePlot.plot(figsize=(20,8), marker='o',color = 'b')
    ax2 = barPlot.plot(kind='bar',ax = ax,rot=1,secondary_y=True,stacked=stacked)
    ax.set_title(df[col].name.title()+' vs Probability Charge Off',fontsize=20,weight="bold")
    ax.set_xlabel(df[col].name.title(),fontsize=14)
    ax.set_ylabel('Probability of Charged off',color = 'b',fontsize=14)
    ax2.set_ylabel('Number of Applicants',color = 'g',fontsize=14)
    plt.show()

# 1. Location vs Probability Charge Off
filter_states = df_loan_cleaned.addr_state.value_counts()
filter_states = filter_states[(filter_states < 10)]
loan_filter_states = df_loan_cleaned.drop(labels = df_loan_cleaned[df_loan_cleaned.addr_state.isin(filter_states.index)].index)
states = crosstab(loan_filter_states,'addr_state')
display(states.tail(20))
bivariate_prob(df =loan_filter_states,col ='addr_state')



#2. Purpose of Loan vs Probability Charge Off
purpose = crosstab(df_loan_cleaned,'purpose')
display(purpose)
bivariate_prob(df =df_loan_cleaned,col ='purpose',stacked=False)



# 3. Annual Income Range vs Probability Charge Off
annual_inc_range = crosstab(df_loan_cleaned,'annual_inc_range')
display(annual_inc_range)
bivariate_prob(df =df_loan_cleaned,col ='annual_inc_range')



# 4. Interest rate Range vs Probability Charge Off
int_rate_range = crosstab(df_loan_cleaned,'int_rate_range')
display(int_rate_range)
bivariate_prob(df =df_loan_cleaned,col ='int_rate_range')


# 5. Employment Length vs Probability Charge Off
emp_length = crosstab(df_loan_cleaned,'emp_length')
display(emp_length)
bivariate_prob(df =df_loan_cleaned,col ='emp_length')


```

## 3.3 PREDICTIVE ANALYSIS

### Goal - To Build a predictive model that will allow an investor to make well informed decisions on what loans to invest in so as to maximize their returns. 

Since the primary goal is to ensure that we avoid loan defaults, it is of paramount importance that we avoid misclassifying bad loans as good (an important point to note here is that classification of a good loan as bad, while reducing the potential pool for investment, will not have the same ramifications).

Data Pre-processing for Predictive Modeling

   - It has an effective method for estimating missing data and maintains accuracy when a large proportion of the data are missing.

# 3.3.b Logistic Regression with Grid Search
Logistic Regression is a form of regression that allows the prediction of discrete variables by a mix of continuous and discrete predicators. 

When we implement Logistic Regression with grid search we got the best accuracies and minimum false negatives.

Here we will use a grid search with cross validation to optimize on the hyperparameters of the algorithm.

- Logistic regression is often used because the relationship between discrete variable(DV) and predicator is non-linear. Instead of using Y (or p) as the dependent variable, we use a function of it, which is called logit. 
- The key advantage of logits maps any value of the dependent variables into a probability [0,1]. The logit, it turns out, can be modelled as a linear function of predicators. Once the logit has been predicted, it can be mapped back to a probability p.   
- Logistic Regression when implemented with grid search we got the best accuracies and minimum false negatives. We used the randomized search to find the best hyper paramters for the model.

```python
from sklearn.model_selection import GridSearchCV

def cross_validation_best_parameters(model, param_grid):
    grid = GridSearchCV(model, param_grid,cv=10, scoring='accuracy')
    X=data_clean.iloc[:,:-1].values
    y=data_clean.iloc[:,-1].values
    grid.fit(X,y)
    mean_scores = [result.mean_validation_score for result in grid.grid_scores_]
    return mean_scores,grid.best_score_,grid.best_estimator_
logreg = linear_model.LogisticRegression(random_state=0)
c = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(C = c)
mean_scores,Best_Accuracy, Best_classifier = cross_validation_best_parameters(logreg,param_grid)
print("Best accuracy is " + str(Best_Accuracy))
print(Best_classifier)


# Display
clf_LR = linear_model.LogisticRegression(C=Best_classifier.C)
clf_LR.fit(X_train,y_train)
LR_Predict = clf_LR.predict_proba(X_test)[:,1]
LR_Predict_bin = clf_LR.predict(X_test)
LR_Accuracy = accuracy_score(y_test,LR_Predict.round())
print("Logistic regression accuracy is ",LR_Accuracy)
plotAUC(y_test,rfPredictproba, 'Random Forest')
plotAUC(y_test,LR_Predict,'Logistic Regression')
plt.show()
plt.figure(figsize = (6,6))
plot_confusion_matrix(LR_Predict_bin, normalize = True)
plt.show()
```
